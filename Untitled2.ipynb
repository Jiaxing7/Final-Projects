```python
df = df.dropna(subset=['title'])
```

```python
df.isna().sum()
```

#### deal with publieshed_date

```python
df['publieshed_date'] = df['publieshed_date'].str.split('shed',expand=True)[1]
```

```python
df['publieshed_date'].str.split(' ', expand=True)
```

```python
# Create a new column for year
df['published_year'] = df['publieshed_date'].str.split(' ', expand=True)[3]
df['published_year'] = pd.to_numeric(df['published_year'], errors='coerce')

# Create a new column for month
df['published_month'] = df['publieshed_date'].str.split(' ', expand=True)[1]

month_dict = {'January': 1, 'February': 2, 'March': 3, 'April': 4, 'May': 5, 'June': 6,
              'July': 7, 'August': 8, 'September': 9, 'October': 10, 'November': 11, 'December': 12}
df['published_month'] = df['published_month'].fillna('0').map(month_dict)

# Create a new column for day
df['published_day'] = df['publieshed_date'].str.split(' ', expand=True)[2]
df['published_day'] = df['published_day'].str.rstrip(',')
df['published_day'] = pd.to_numeric(df['published_day'], errors='coerce')
```

```python
df.isna().sum()
```

```python
df[df['published_year'].isna()].head(60)
```

#### deal with rating_n, review_n

```python
df['rating_n'] = df['rating_n'].str.split('r',expand=True)[0]
df['rating_n'] = df['rating_n'].fillna('0').str.replace(',', '').astype(int)
```

```python
df['review_n'] = df['review_n'].str.split('r',expand=True)[0]
df['review_n'] = df['review_n'].fillna('0').str.replace(',', '').astype(int)
```

#### deal with author_followers

```python
if 'k' in df['author_followers']:
    df['author_followers'] = df['author_followers'].str.replace('k','')
    df['author_followers'] = pd.to_numeric(df['author_followers'])*1000
# else:
#     df['author_followers'] = df['author_followers'].fillna('0').str.replace(',','').astype(float)
```
