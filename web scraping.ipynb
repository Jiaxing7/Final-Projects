```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
from time import sleep
```

# Web scraping

```python
iterations = range(1, 4)   
books = []

 
for i in iterations:
    start_at= str(i)
    url = "https://www.goodreads.com/list/show/183940.Best_books_of_2023" + start_at
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    for b in soup.select("td:nth-child(3) > a"):
        link = b.get("href")    
        if link is not None:
            books.append(b["href"])
    
```

```python
len(books)
```

```python
# 2. find url and store it in a variable

book_soups = []

for b in books:
    # send request
    url = "https://www.goodreads.com" + b
    response = requests.get(url)
    print(b, response.status_code)

    # parse & store html
    soup = BeautifulSoup(response.content, "html.parser")
    book_soups.append(soup.select("div.BookPage__mainContent"))

    # respectful nap:
    wait_time = random.randint(1,2)
    print("I will sleep for " + str(wait_time) + " second/s.")
    sleep(wait_time)
```

```python
len(book_soups)
```

```python
title = []
author = []
publieshed_date = []
rating = []
rating_n = []
review_n = []
introduction = []
genre = []
page = []
books_written_n = []
author_followers = []
link = []

for i in range(len(book_soups)):
    try:
        title.append(book_soups[i][0].select('h1')[0].text.strip())
    except:   
        title.append("NA")
    try:
        author.append(book_soups[i][0].select('span.ContributorLink__name')[0].text.strip())   
    except:
        author.append("NA")
    try:
        publieshed_date.append(book_soups[i][0].select('p')[1].text.strip())
    except:
        publieshed_date.append("NA")
    try:
        rating.append(book_soups[i][0].select('div.RatingStatistics__rating')[0].text.strip())
    except:
        rating.append("NA")
    try:
        rating_n.append(book_soups[i][0].select("div.RatingStatistics__column")[1].select('span')[0].text.strip())
    except:
        rating_n.append("NA")
    try:
        review_n.append(book_soups[i][0].select("div.RatingStatistics__column")[1].select('span')[1].text.strip())
    except:
        review_n.append("NA")
    try:
        introduction.append(book_soups[i][0].select('span.Formatted')[0].text.strip())
    except:
        introduction.append("NA")
    try:
        genre.append([tag.text.strip() for tag in book_soups[i][0].select('span.BookPageMetadataSection__genreButton')])
    except:
        genre.append("NA")
    try:
        page.append(book_soups[i][0].select("p")[0].text.strip())
    except:
        page.append("NA")
    try:
        books_written_n.append(book_soups[i][0].select_one('div.FeaturedPerson__infoPrimary span.Text__body3').text.strip().split("\xa0")[0].strip())
    except:
        books_written_n.append("NA")
    try:
        author_followers.append(book_soups[i][0].select('span.u-dot-before')[1].text.strip().split("followers")[0].strip())
    except:
        author_followers.append("NA")
        

for i in range(len(books)):
    link.append("https://www.goodreads.com" + books[i])
    
df1 = pd.DataFrame({"title":title,
                   "author":author,
                   "publieshed_date": publieshed_date,
                   "rating": rating,
                   "rating_n": rating_n,
                   "review_n": review_n,
                   "introduction": introduction,
                   "genre": genre,
                   "page": page,
                   "books_written_n": books_written_n,
                   "author_followers": author_followers,
                   "link":link
                  })
```
